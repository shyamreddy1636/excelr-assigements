```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"D:\Assignments\EDA1\EDA1\Cardiotocographic.csv")
data
```

change the raw data into the data frame


```python
df=pd.DataFrame(data)
```

count the all null values in the data set

```python
df.isnull().sum()
```

fill the null values in the data set with mean value and "inplace=True" replaces the all null values

```python
df['LB'].fillna(df.LB.mean(),inplace=True)
df['AC'].fillna(df.AC.mean(),inplace=True)
df['DS'].fillna(df.DS.mean(),inplace=True)
df['DP'].fillna(df.DP.mean(),inplace=True)
df['MLTV'].fillna(df.MLTV.mean(),inplace=True)
df['Width'].fillna(df.Width.mean(),inplace=True)
df['Tendency'].fillna(df.Tendency.mean(),inplace=True)
df['NSP'].fillna(df.NSP.mean(),inplace=True)


```

Box plot is used to identify the out layers

```python
fig, box=plt.subplots(5,3, figsize=(11,26))

(df['LB']).plot(kind='box', ax=box[0,0])
box[0,0].set_title('LB')

(df['AC']).plot(kind='box', ax=box[0,1])
box[0,1].set_title('AC')

(df['FM']).plot(kind='box', ax=box[0,2])
box[0,2].set_title('FM')

(df['UC']).plot(kind='box', ax=box[1,0])
box[1,0].set_title('UC')

(df['DL']).plot(kind='box', ax=box[1,1])
box[1,1].set_title('DL')

(df['DS']).plot(kind='box', ax=box[1,2])
box[1,2].set_title('DS')

(df['DP']).plot(kind='box', ax=box[2,0])
box[2,0].set_title('DP')

(df['ASTV']).plot(kind='box', ax=box[2,1])
box[2,1].set_title('ASTV')

(df['MSTV']).plot(kind='box', ax=box[2,2])
box[2,2].set_title('MSTV')

(df['ALTV']).plot(kind='box', ax=box[3,0])
box[3,0].set_title('ALTV')

(df['MLTV']).plot(kind='box', ax=box[3,1])
box[3,1].set_title('MLTV')

(df['Width']).plot(kind='box', ax=box[3,2])
box[3,2].set_title('Width')

(df['Tendency']).plot(kind='box', ax=box[4,0])
box[4,0].set_title('Tendency')

(df['NSP']).plot(kind='box', ax=box[4,1])
box[4,1].set_title('NSP')

fig.tight_layout()
plt.show()
```
