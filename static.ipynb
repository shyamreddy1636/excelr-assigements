```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\Assignments\Basic stats - 1\sales_data_with_discounts.csv')
df
```

To get the all numericel values in data set:

```python
num=df.select_dtypes(include=np.number)
num
```

Descriptive Analytics for Numerical Columns:


            *  finding mean,mode,median and standard devation by using for loop

```python
mean=[]
mode=[]
median=[]
std=[]

for i in num:
    mean.append(num[i].mean())
    median.append(num[i].median())
    std.append(num[i].std())
    mode.append(num[i].mode())



print('mean is:',mean) 
print('median is',median)   
print('standard deviation  is:',std)
print('mode is:',mode)
```

Histogram:

it is used on continues variables or continues data

```python
fig, hist=plt.subplots(3,2, figsize=(16,10))
(df['Volume']).plot(kind='hist', ax=hist[0,0])
hist[0,0].set_title('Volume')

(df['Total Sales Value']).plot(kind='hist', ax=hist[0,1])
hist[0,1].set_title('Total Sales Value')

(df['Avg Price']).plot(kind='hist', ax=hist[1,0])
hist[1,0].set_title('Avg Price')


(df['Discount Amount']).plot(kind='hist', ax=hist[1,1])
hist[1,1].set_title('Discount Amount')

(df['Net Sales Value']).plot(kind='hist', ax=hist[2,0])
hist[2,0].set_title('Net Sales Value')
fig.tight_layout()
plt.show()
```
